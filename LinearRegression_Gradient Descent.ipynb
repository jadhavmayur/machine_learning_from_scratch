## üìò Linear Regression from Scratch ‚Äì Summary

### üßÆ Goal

To find the best-fit line using the equation:

$$
\hat{Y} = mX + c
$$

Where:
- $\hat{Y}$: Predicted value  
- $X$: Input feature  
- $m$: Slope of the line  
- $c$: Intercept  

---

### üìâ Loss Function (Mean Squared Error)

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2
$$

Where:
- $Y_i$: Actual value  
- $\hat{Y}_i = mX_i + c$: Predicted value  
- $n$: Number of data points  

---

### üîÅ Gradient Descent

We minimize the loss function by updating $m$ and $c$ using gradient descent.

#### üßÆ Derivatives

- Derivative of loss with respect to **$m$**:

$$
\frac{\partial \text{MSE}}{\partial m} = -\frac{2}{n} \sum_{i=1}^{n} X_i (Y_i - \hat{Y}_i)
$$

- Derivative of loss with respect to **$c$**:

$$
\frac{\partial \text{MSE}}{\partial c} = -\frac{2}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)
$$

---

### üîÑ Parameter Update Rules

$$
m = m - \alpha \cdot \frac{\partial \text{MSE}}{\partial m}
$$

$$
c = c - \alpha \cdot \frac{\partial \text{MSE}}{\partial c}
$$

Where:
- $\alpha$: Learning rate

---

### ‚úÖ Steps Summary

1. Initialize $m = 0$, $c = 0$  
2. For each epoch:
   - Compute predictions: $\hat{Y} = mX + c$
   - Calculate gradients
   - Update $m$ and $c$
3. Plot the final regression line

---


### Importing necessary libraries


```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Mounting Google Drive to access data stored there
from google.colab import drive
drive.mount("/content/drive")
# Setting default figure size for plots
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

### Data Loading and Preprocessing

```python
# Reading the dataset from Google Drive
data = pd.read_csv('/content/drive/MyDrive/ml_scratch/data.csv')
# Separating input (X) and output (Y) variables
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
# Visualizing the data points
plt.scatter(X, Y)
plt.show()
```

```python
# Initializing model parameters
m = 0  # slope
c = 0  # y-intercept

# Setting the learning rate and number of iterations (epochs)
L = 0.0001
epochs = 1000

# Number of data points
n = float(len(X))

# Gradient Descent loop
for i in range(epochs):
    # Calculating the predicted Y values
    Y_pred = (m * X) + c

    # Calculating the gradients (partial derivatives)
    D_m = -(2/n) * sum(X * (Y - Y_pred))  # derivative w.r.t. m
    D_c = -(2/n) * sum(Y - Y_pred)        # derivative w.r.t. c

    # Updating the parameters using the gradients
    m = m - (L * D_m)
    c = c - (L * D_c)

# Printing the final values of m and c
print(m, c)
```

```python
# Generating predicted values using the learned parameters
Y_pred = m * X + c
```

```python
# Plotting the original data points
plt.scatter(X, Y)

# Plotting the regression line
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')
plt.show()
```
